```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataza.csv')
df.head()
```

```python

df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Location', hue='Indicator', palette='Set1')
plt.title('Counts of Different Indicators by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Indicator', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Create a stacked bar plot for indicators by location
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='Location', hue='Indicator', multiple='stack', palette='Set1')
plt.title('Stacked Bar Plot of Indicators by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Indicator', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 3. Calculate and display average values by location and indicator
avg_values = df.groupby(['Location', 'Indicator'])['Value'].mean().unstack()
print("Average Values by Location and Indicator:")
print(avg_values)
```

```python
# 4. Identify locations with the highest and lowest average values for a specific indicator
indicator_of_interest = 'Total fertility rate 15-49'
avg_values_for_indicator = avg_values[indicator_of_interest]
highest_location = avg_values_for_indicator.idxmax()
lowest_location = avg_values_for_indicator.idxmin()
print(f"For '{indicator_of_interest}':")
print(f"Highest Average Value Location: {highest_location} ({avg_values_for_indicator.max():.2f})")
print(f"Lowest Average Value Location: {lowest_location} ({avg_values_for_indicator.min():.2f})")
```
